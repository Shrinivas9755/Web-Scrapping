```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://insights.blackcoffer.com/rising-it-cities-and-its-impact-on-the-economy-environment-infrastructure-and-city-life-by-the-year-2040-2/'
```

```python
#requesting to url
response = requests.get(url)
```

```python
#article text and title extraction
title = soup.find('h1').text.strip()
article_text = ' '
for paragraph in soup.find_all('div', class_ = 'td-post-content'):
    article_text += paragraph.text.strip()
```

```python
#article text and title saving to a text file
filename = "URL_ID 01.txt"
with open(filename, 'w', encoding = 'utf-8') as file:
    file.write(title + "\n\n")
    file.write(article_text)
print("Article saved to", filename)
```

```python
import nltk
nltk.download('punkt')
import pyphen
from textblob import TextBlob
import spacy
from openpyxl import load_workbook

nlp = spacy.load("en_core_web_sm")
```

```python
#defining common suffixes
complex_suffixes = set(["able", "ible", "al", "ial", "ed", "en", "er", "est", "ful", "ic", "ing", "ion", "tion", "ation", "ition", "ity", "ty", "ive", "ative", "itive", "less", "ly", "ment", "ness", "ous", "eous", "ious", "s", "es", "y"])
```

```python
#defining list of names for processing
file_list = [
    "URL_ID 01.txt","URL_ID 02.txt","URL_ID 03.txt","URL_ID 04.txt","URL_ID 05.txt","URL_ID 06.txt","URL_ID 07.txt","URL_ID 08.txt","URL_ID 09.txt","URL_ID 10.txt"]
```

```python
#creating empty list
result_dfs = []
```

```python
#using for loops performing analysis on each file
for filename in file_list:
    #loading text article
    with open (filename, 'r', encoding="UTF-8") as file:
        article_text = file.read()

        #tokenising text into sentences and words
        sentences = nltk.sent_tokenize(article_text)
        words = nltk.word_tokenize(article_text)

        #initialise syllable counter
        dic = pyphen.Pyphen(lang='en')

        #sentiment analysis using textblob
        analysis = TextBlob(article_text)
        positive_score = analysis.sentiment.polarity
        negative_score = analysis.sentiment.polarity * -1

        #linguistic features using spacy
        doc = nlp(article_text)
        personal_pronouns = sum([1 for token in doc if token.pos_ == "PRON" and token.dep_ =="nsubj"])

        #fog index
        words = nltk.word_tokenize(article_text)
        complex_word_count = sum([1 for word in words if any(suffix in word[-6:] for suffix in complex_suffixes)])


        #additional features
        subjectivity_score = analysis.sentiment.subjectivity
        avg_number_of_words_per_sentence = len(words) / len(nltk.sent_tokenize(article_text))

         #calculating various token
        word_count = len(words)
        sentence_count = len(sentences)
        syllable_count = sum([len(dic.inserted(word).split("-")) for word in words])
        avg_sentence_length = word_count / sentence_count
        avg_word_length = syllable_count / word_count
        complex_word_count = sum([1 for word in words if any(suffix in word[-6:]for suffix in complex_suffixes)])
        complex_word_percentage = complex_word_count / word_count * 100
        fog_index = 0.1 * (avg_sentence_length + complex_word_percentage)
       #creating dataframe with the results
        df = pd.DataFrame({
            'Filename': [filename],
            'POSITIVE SCORE':[positive_score],
            'NEGATIVE SCORE':[negative_score],
            'POLARITY SCORE':[analysis.sentiment.subjectivity],
            'SUBJECTIVITY SCORE':[subjectivity_score],
            'Avg Sentence Length': [avg_sentence_length],
            'PERCENTAGE OF COMPLEX WORDS': [complex_word_percentage],
            'FOG INDEX':[fog_index],
            'AVG NUMBER OF WORDS PER SENTENCE':[avg_number_of_words_per_sentence],
            'Complex Word Count': [complex_word_count],
            'Word Count': [word_count],
            'Syllable Count': [syllable_count],
            'PERSONAL PRONOUNS':[personal_pronouns],
            'Avg Word Length': [avg_word_length],
        })

        #appending the results
        result_dfs.append(df)

    #concatinating all the dataframe into one 
    result_df = pd.concat(result_dfs, ignore_index=True)

    with pd.ExcelWriter(r'D:\Data Science\output_final.xlsx') as writer:
        result_df.to_excel(writer, index=False)

print(result_df)

    
```
